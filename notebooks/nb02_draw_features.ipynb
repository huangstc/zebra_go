### Draw Training Example

This notebook draws feature planes of a training example.

```python
import numpy as np
import tensorflow as tf

from visualization import SvgGoBoard
from train_utils import LabelToCoord, ParseForPolicy, DrawGoBoard

tf.enable_eager_execution()
```

```python
def ReadExamplesFromTFRecords(filename, max_num_examples=1000):
    """ Reads the first N examples from a TFRecord file.
    """
    # The file is compressed, by ../model/gen_dataset
    dataset = tf.data.TFRecordDataset([filename], 'ZLIB')
    examples = []
    examples_in_proto = []
    
    for ex in dataset:
        examples.append(ex)
        examples_in_proto.append(tf.train.Example.FromString(ex.numpy()))
        
    return examples, examples_in_proto
```

```python
moves, _ = ReadExamplesFromTFRecords('shusai_19000415.rio')
```

```python
planes, label = ParseForPolicy(moves[220])

# The orginal Go board state.
orig = planes.numpy()[:,:,0]

# Current player's chains with 1, 2 or 3 liberties.
b_planes = [planes.numpy()[:,:,k] for k in [1,2,3]]

# The opponent's chains with 1, 2, or 3 liberties.
w_planes = [-1*planes.numpy()[:,:,k] for k in [4,5,6]]

# Next move for current player:
next_move = LabelToCoord(label.numpy())
```

```python
bb = DrawGoBoard(orig)
bb.AddSquare(next_move[0], next_move[1], 0)
bb
```
